## Introducción
Hasta este punto, nosotros ya conocemos las herramientas necesarias para empezar a hacer análisis de datos: un lenguaje de programación potente, un conocimiento sobre matrices y sus operaciones, una herramienta para la manipulación de datos y otra herramienta para visualizar y hacer gráficas potentes y personalizables. ## Pasos para realizar un análisis de datos
Al igual que en los métodos de investigación, se pueden aplicar algoritmos o estrategias para realizar la actividad del análisis de datos. Estas estrategias pueden variar según la aproximación que cada quién le quiera dar, no obstante, siempre se siguen los siguientes pasos u objetivos.

### Paso 1: Plantear una o varias cuestiones
Dependiendo del contexto, antes de empezar cualquier análisis, es necesario tener un primer objetivo en mente. A partir de este objetivo uno puede plantearse ciertas preguntas que ayuden a darle solidez a los planteamientos.

Por ejemplo, supongamos que somos los dueños de una empresa y queremos potenciar nuestras ventas. En este caso, el objetivo es claro __potenciar nuestras ventas__, ahora, a partir de este objetivo, debemos plantearnos las preguntas necesarias para darle solidez, estos cuestionamientos (para este contexto) pueden ser tales como: "¿Cuál es el rango de edad de los clientes que consume mi producto?", "¿Cuáles son los días en que más se vende?, "¿Cuál es el rango de precios que más le gusta a mis clientes?", etc. 

Es fundamental tener un objetivo y algunos cuestionamientos bien planteados desde el principio, ya que a partir de aquí es cuando vamos a saber qué datos debemos de buscar.

### Paso 2: Recolección de datos
Este puede ser, quizá, el paso más complicado de todos. En este paso debemos de buscar uno o varios datasets que vayan de acuerdo a nuestro objetivo y, por supuesto, satisfagan parcial o totalmente las cuestiones que planteamos en el paso 1. 

El motivo por el que este paso suele ser complicado es debido a que __no__ siempre se encuentran conjuntos útiles de datos para este propósito. Esto se puede dar por diversos motivos, el más común es, casi siempre, porque los datos son privados, aunque puede darse el caso que nuestras cuestiones son demasiado específicas.

Cuando no es posible encontrar datasets útiles y el objetivo no puede ser modificado, no queda de otra que uno mismo recolectar los datos. Esto se puede lograr a través de encuestas, extrapolación de bases de datos, utilizar herramientas de recolección de datos tales como Google Analytics, Facebook Insights, entre otras.

Una vez se tiene un conjunto de datos inicial, podemos proceder con el siguiente paso.

### Paso 3: Limpieza de datos
Este paso es muy importante, aquí lo que se hace es realizar las operaciones necesarias sobre los datos recolectados, con el fin de explorar y facilitar el análisis de estos. 

Este paso suele ser variado en su complejidad debido a que depende mucho de la naturaleza de los datos sobre los que se están trabajando. Por ejemplo, si la gran mayoría de nuestros datos son cadenas de texto, es muy posible que tengamos que aplicar técnicas de tratamiento de lenguaje natural, codificación de texto, limpieza de basura, corrección de ortografía, normalización del texto, etc. En caso que sean numéricos casi siempre hay que aplicar normalizaciones de escala, eliminación de ruido, reemplazo de nulidades, entre otros. 

_Generalmente los datos numéricos son más sencillos de tratar._

La mayoría de las técnicas de limpieza se aprenden con la práctica y la experiencia, dado que cada conjuto de datos es potencialmente diferente entre sí.

El _ultimate goal_ de este paso es que los datos estén lo suficientemente _pulidos_ para que podamos analizarlos cómodamente.

### Paso 4: Analizar los datos
Este paso suele ser divertido. Aquí lo que haremos será aplicar diversas operaciones, dependientes del contexto, para encontrar patrones, correlaciones, información detallada, etc. Es en este paso en donde se aplica ingenio para poder sacarle el máximo provecho a los datos que con esfuerzo nos encargamos de pulir en el paso anterior. Muchas veces se plantean hipótesis sobre cada una de las cuestiones planteadas en el paso 1, con el fin de explorar diversas posibilidades.

En este paso también nos encargaremos de generar gráficas que nos ayuden a entender el comportamiento de los datos, esto no solo nos apoya en resolver los planteamientos, sino que también brinda soporte a el paso siguiente.

La calidad del análisis va a depender muchísimo de la calidad de los datos, por lo tanto, es crucial que el paso 2 y 3 hayan sido muy efectivos.

### Paso 5: Interpretando los resultados y generación de conclusiones
Este es el paso final. Aquí lo que haremos será, una vez hecho un buen análisis, agrupar nuestros resultados y plantear conclusiones que vayan de acuerdo a estos. Aquí haremos mucho uso de las gráficas que hayamos, probablemente, hecho en el paso anterior.

_Unas buenas conclusiones siempre van a ir acompañadas información que las respalden y, de ser posible, una o varias gráficas que reafirmen lo estipulado_

No siempre las conclusiones van a responder de forma positiva a nuestro objetivo, muchas veces puede que incluso lo contradigan. ¡Esto no significa que esté mal! Sin haber hecho un análisis de datos, puede que hayamos caído en posibles sesgos que llevaran a tomar decisiones incorrectas.

### 4. Análisis
Ya tenemos un buen conjunto de datos fiables con los cuales podemos empezar a analizar, plantear hipótesis, ver correlaciones, etc.

Recordemos los planteamientos:
- ¿Influye en algo si el estudiante es de colegio público o privado?
- ¿Los estudiantes con internet tienen mejores resultados?

Empecemos viendo si hay algún tipo de correlación entre los colegios públicos y los puntajes.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

Para esto, haremos una pequeña gráfica de dispersión, para revisar, con un simple vistazo, el comportamiento de los datos.

```python
plt.scatter(datos_ini.COLE_NATURALEZA, datos_ini.PUNT_GLOBAL);
```

Según la gráfica, podemos hacer la hipótesis que los colegios privados (no oficiales), tienen una ligera ventaja sobre los públicos. Veamos más numéricamente esto, aplicando filtros y viendo sus datos estadísticos, haciendo uso de la función describe.

Empecemos por los colegios privados y luego los públicos.

```python
PRIV = datos_ini[datos_ini.COLE_NATURALEZA == "NO OFICIAL"].describe()
PRIV
```

```python
PUB = datos_ini[datos_ini.COLE_NATURALEZA == "OFICIAL"].describe()
PUB
```

Podemos ver que hay una diferencia notable entre los puntajes de los colegios privados sobre los públicos. Específicamente, la ventaja porcentual sobre el puntaje global es:

```python
# Nótese que acá podríamos usar el método mean de numpy
MEDIA_PUNT_OF = PUB.PUNT_GLOBAL[1]
MEDIA_PUNT_PRIV = PRIV.PUNT_GLOBAL[1]
VENTAJA_PRIV = (MEDIA_PUNT_PRIV/MEDIA_PUNT_OF-1)*100
VENTAJA_PRIV
```

Hagamos algo más interesante, una gráfica de barras que compare los valores.

```python
# Vamos a tomar como el eje X, es decir, las categorías de las barras, como las columnas del DataFrame del describe
eje_x = PRIV.columns[:5]
# Dado que son 2 barras, una para los colegios privados y otra para públicos tendremos que tomar el valor de media para ambos
altura_priv = PRIV.iloc[1,:5] # Segunda fila (mean) de todas las columnas, excepto la última
altura_pub = PUB.iloc[1, :5]
# Ya tenemos el eje X y las alturas. Vamos a hacer la figura.
grosor = 0.25 # El grosor que tendrá cada rectángulo
barras_pub, ax = plt.subplots(figsize=(20,10))
ax.set_title("Comparación de puntajes entre colegios públicos y privados pruebas ICFES")
ax.set_ylabel("Puntaje promedio")
ax.set_xlabel("Área del conocimiento")
bpriv = ax.bar(eje_x, altura_priv, label = "Colegios privados", width=grosor)
bpub = ax.bar(eje_x, altura_pub, color = "r", label = "Colegios públicos", width=grosor)
ax.legend();
for tipo in [bpriv, bpub]:
    for barra in tipo:
        alto = barra.get_height() # Vamos a tomar la altura de la barra, ya que este es el valor real que tiene.
        alto = np.around(alto, 2) # Redondeamos el valor a 2 decimales
        ancho = barra.get_width() # Tomamos el valor del ancho del rectángulo.
        x = barra.get_x() + ancho # Queremos que la coordenada x del texto esté al lado del rectángulo.
        y = alto # Queremos que la coordenada y del texto esté a la altura del rectángulo.
        ax.text(x,y, alto)
```

Ya tenemos la primera gráfica interesante para apoyar nuestro planteamiento inicial. Pasemos al siguiente planteamiento.

Veamos si hay alguna correlación entre tener internet y el puntaje global. Aplicaremos una estrategia similar al planteamiento anterior.

```python
plt.scatter(datos_ini.FAMI_TIENEINTERNET, datos_ini.PUNT_GLOBAL);
```

Aparentemente, los estudiantes que tienen acceso a Internet tienen unos resultados ligeramente superiores que los que no. Veamos si, por área de estudio, influye.

```python
columnas = datos_ini.columns[5:-1] # Tomaremos los nombres de las columnas, a partir de la que tiene el id 5 hasta la penúltima
for columna in columnas:
    fig, ax = plt.subplots()
    ax.set_title(columna)
    ax.scatter(datos_ini.FAMI_TIENEINTERNET, datos_ini[columna]);
```

Gráficamente  no parece afectar en gran medida, en especial para la puntuación del área de matemáticas. Veamos los números para tener una mayor precisión.

```python
INT = datos_ini[datos_ini.FAMI_TIENEINTERNET == "Si"].describe()
INT
```

```python
NO_INT = datos_ini[datos_ini.FAMI_TIENEINTERNET == "No"].describe()
NO_INT
```

Al final, sí que hay una relación entre ambos. Los estudiantes con acceso a internet tienen mejores puntajes globales. La mejora porcentual sobre los puntajes globales es de:

```python
VENTAJA_INT = (INT.PUNT_GLOBAL[1]/NO_INT.PUNT_GLOBAL[1]-1)*100
VENTAJA_INT
```

Para no repetir una gráfica similar a la anterior, mostremos la cantidad porcentual de estudiantes con y sin internet:

```python
figura_int, ax = plt.subplots(dpi=90) # dpi es un parámetro que sube la densidad de puntos por pulgada (resolución)
ax.set_title("Porcentajes de estudiantes con acceso a internet")
ax.pie(datos_ini.FAMI_TIENEINTERNET.value_counts(sort=False), labels = datos_ini.FAMI_TIENEINTERNET.unique(), startangle=90, autopct="%1.1f%%");
```

Con estos datos que responden y respaldan nuestros planteamientos, procedamos al último paso.

Vamos a hacer las conclusiones. Retomemos los datos que utilizaríamos.

```python
print (VENTAJA_PRIV) # Ventajas de ser estudiante de colegio privado
print (VENTAJA_INT) # Ventajas de tener internet
```
